```python
import pandas as pd 
import numpy as np 
import joblib
```

**First Phase Data**

```python
numbers = ['0500','1000','1500','2000','2500','3000','3500','4000','4500','5000','5500']
files = []
for num in numbers: 
    file = pd.read_csv(f'data/yahoo_finance/financials_raw{num}')
    files.append(file)


financials_raw = pd.concat(files)
financials_raw.info()
```

```python
financials_raw.drop(financials_raw.columns[np.r_[0,2,15,16,18,19,20]],inplace=True, axis=1)
```

```python
financials_raw.info()
```

**Second Phase Data**

```python
numbers = ['0500','1000','1500','2000','2500','3000','3500','4000','4500','5000',
           '5500','5500','6000','6500','7000','7500','8000','9000','9500', '10000','10500','11000']

files = []
for num in numbers: 
    file = pd.read_csv(f'data/yahoo_finance/yf1_{num}')
    files.append(file)


yf_raw = pd.concat(files)
yf_raw.info()
 

```

```python
yf_raw.drop(yf_raw.columns[np.r_[0]],inplace=True, axis=1)
```

```python
yf_raw.info()
```

**Combine first and second phase data**

```python
df12 = [yf_raw,financials_raw]
yf1 = pd.concat(df12)
yf1.info()
```

```python
yf1.rename(columns={
    'symbol':'ticker',
    'totalRevenue':'total_revenue',
    'fullTimeEmployees': 'full_time_employees',
    'ebitdaMargins': 'ebitda_margins',
    'profitMargins':'profit_margins',
    'grossMargins': 'gross_margins',
    'operatingCashflow': 'operating_cashflow',
    'revenueGrowth':'revenue_growth',
    'operatingMargins': 'operating_margins',
    'grossProfits': 'gross_profits',
    'bookValue': 'book_value',    
    'netIncomeToCommon': 'net_income_to_common',
    'enterpriseToRevenue': 'enterprise_to_revenue',
    'enterpriseValue': 'enterprise_value',
    'marketCap': 'market_cap'
},
                inplace=True)
```

```python
yf1 = yf1[yf1['ticker']!='Null']
```

```python
def to_float(df,col):        
    df[col] = df[col].replace('Null',0).astype(float)
    return df[col]

def to_int(df,col):       
    df[col] = df[col].fillna(0).replace('Null',0).astype(int)
    return df[col]
```

```python
floats=['ebitda_margins',
        'profit_margins',
        'gross_margins',
        'operating_margins',
        'revenue_growth',
        'book_value',
        'enterprise_to_revenue',
       'price_to_book']

ints=['total_revenue',
      'operating_cashflow',
      'full_time_employees',
      'ebitda',
      'gross_profits',
      'net_income_to_common',
      'enterprise_value',
      'market_cap',
     'shares_short']

for col in list(yf1.columns):
    if col in floats: 
        to_float(yf1,col)
        
    elif col in ints: 
        to_int(yf1,col)
    else:
        None
        
```

```python
yf1.drop(columns='nextFiscalYearEnd',inplace=True,axis=1)
```
